```python
import csv
from pathlib import Path
import pandas as pd
from pyprojroot.here import here
```

```python
path = here('data/gtex/GTEx_Analysis_v8_Annotations_SampleAttributesDS.txt')

gtex_meta = pd.read_csv(
    path,
    sep='\t',
    header=0,
    dtype=str,
    quoting=csv.QUOTE_NONE,   
    engine='python',
    comment=None,            
    keep_default_na=False,
    on_bad_lines='warn'     
)

print(gtex_meta.shape)
gtex_meta.head()
```

```python
gtex_archs4_projection_filtered = pd.read_csv(
    here("output/archs4/gtex_archs4_projection_filtered.csv"),
    index_col=0
)

gtex_archs4_projection_filtered.head()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering
from sklearn.metrics import (
    silhouette_score, silhouette_samples,
    adjusted_rand_score, normalized_mutual_info_score,
    confusion_matrix
)

# ---------------- Align B (LVs x samples) with metadata ----------------
X = gtex_archs4_projection_filtered.T.copy()   # samples x LVs
X.index.name = "sample_id"

meta = gtex_meta.loc[:, ["SAMPID", "SMTS"]].copy()
meta["SAMPID"] = meta["SAMPID"].astype(str).str.strip()

common = X.index.intersection(meta["SAMPID"])
if len(common) < 5:
    raise ValueError("Too few overlapping samples between B and metadata.")

X = X.loc[common]
meta = meta.set_index("SAMPID").loc[common].reset_index().rename(columns={"SAMPID":"sample_id", "SMTS":"tissue"})
tissues = pd.Categorical(meta["tissue"])
y_true = tissues.codes

# ---------------- Scale features ----------------
scaler = StandardScaler()
Xs = scaler.fit_transform(X.values)

# ---------------- PCA for clustering + PCA(2D) for plotting ----------------
np.random.seed(123)
pca_full = PCA(n_components=min(50, Xs.shape[1], Xs.shape[0]-1), random_state=123)
Z = pca_full.fit_transform(Xs)
var80_n = max(2, (np.cumsum(pca_full.explained_variance_ratio_) >= 0.80).argmax() + 1)

pca2 = PCA(n_components=2, random_state=123)
Z2 = pca2.fit_transform(Xs)
pca_df = pd.DataFrame({"sample_id": X.index, "PC1": Z2[:,0], "PC2": Z2[:,1], "tissue": tissues.astype(str)})

# ---------------- t-SNE(2D) for plotting (version-safe) ----------------
from inspect import signature
tsne_kwargs = dict(n_components=2, perplexity=30, init="pca", random_state=123, verbose=1)
if "n_iter" in signature(TSNE).parameters:
    tsne_kwargs["n_iter"] = 1000
elif "max_iter" in signature(TSNE).parameters:
    tsne_kwargs["max_iter"] = 1000
# Older sklearns don't accept "auto" for learning_rate; use numeric for max compat
if "learning_rate" in signature(TSNE).parameters:
    tsne_kwargs["learning_rate"] = 200

tsne = TSNE(**tsne_kwargs)
Y = tsne.fit_transform(Xs)
tsne_df = pd.DataFrame({"sample_id": X.index, "tSNE_1": Y[:,0], "tSNE_2": Y[:,1], "tissue": tissues.astype(str)})

# ---------------- Label-based silhouette (do tissues form clusters?) ----------------
sil_lbl_overall = np.nan
sil_lbl_samples = np.full(Xs.shape[0], np.nan)
if len(np.unique(y_true)) >= 2:
    try:
        sil_lbl_overall = silhouette_score(Xs, y_true, metric="euclidean")
        sil_lbl_samples = silhouette_samples(Xs, y_true, metric="euclidean")
    except Exception:
        pass
tsne_df["silhouette_label"] = sil_lbl_samples
sil_by_tissue = tsne_df.groupby("tissue", as_index=False)["silhouette_label"].mean()

# ---------------- Unsupervised clustering (k-means, agglomerative) ----------------
K = len(tissues.categories)
km = KMeans(n_clusters=K, n_init=20, random_state=123)
y_km = km.fit_predict(Z[:, :var80_n])

ag = AgglomerativeClustering(n_clusters=K, linkage="ward")
y_ag = ag.fit_predict(Z[:, :var80_n])

# ---------------- Metrics vs tissue ----------------
cm_km = confusion_matrix(y_true, y_km)
ari_km = adjusted_rand_score(y_true, y_km)
nmi_km = normalized_mutual_info_score(y_true, y_km)
pur_km = cm_km.max(axis=0).sum() / cm_km.sum()
sil_km_overall = np.nan
try:
    sil_km_overall = silhouette_score(Xs, y_km, metric="euclidean") if len(np.unique(y_km)) > 1 else np.nan
except Exception:
    pass
map_km = {c: np.argmax(cm_km[:, j]) for j, c in enumerate(range(cm_km.shape[1]))}
acc_km = (np.array([map_km[c] for c in y_km]) == y_true).mean()

cm_ag = confusion_matrix(y_true, y_ag)
ari_ag = adjusted_rand_score(y_true, y_ag)
nmi_ag = normalized_mutual_info_score(y_true, y_ag)
pur_ag = cm_ag.max(axis=0).sum() / cm_ag.sum()
sil_ag_overall = np.nan
try:
    sil_ag_overall = silhouette_score(Xs, y_ag, metric="euclidean") if len(np.unique(y_ag)) > 1 else np.nan
except Exception:
    pass
map_ag = {c: np.argmax(cm_ag[:, j]) for j, c in enumerate(range(cm_ag.shape[1]))}
acc_ag = (np.array([map_ag[c] for c in y_ag]) == y_true).mean()

print("Samples:", Xs.shape[0], "| LVs:", Xs.shape[1], "| PCA dims used:", var80_n)
print("\nLabel (tissue) silhouette overall:", np.round(sil_lbl_overall, 3))
print("k-means:   ARI =", np.round(ari_km, 3), "NMI =", np.round(nmi_km, 3), "purity =", np.round(pur_km, 3), "acc =", np.round(acc_km, 3), "sil =", np.round(sil_km_overall, 3))
print("Agglomer.: ARI =", np.round(ari_ag, 3), "NMI =", np.round(nmi_ag, 3), "purity =", np.round(pur_ag, 3), "acc =", np.round(acc_ag, 3), "sil =", np.round(sil_ag_overall, 3))

# ---------------- Plots (no saving) ----------------
codes_tissue = pd.Categorical(pca_df["tissue"]).codes
codes_km = y_km.astype(int)
codes_ag = y_ag.astype(int)
import matplotlib.lines as mlines

# PCA colored by tissue
plt.figure(figsize=(7.2, 6))
sc = plt.scatter(pca_df["PC1"], pca_df["PC2"], c=codes_tissue, s=10, alpha=0.8)
proxies = [mlines.Line2D([], [], marker='o', linestyle='None') for _ in range(len(tissues.categories))]
for i, p in enumerate(proxies):
    p.set_markerfacecolor(sc.cmap(sc.norm(i))); p.set_markeredgecolor('black'); p.set_markersize(6)
plt.legend(proxies, list(tissues.categories), title="Tissue", loc="best", fontsize=8, frameon=True)
plt.title("PCA (2D) colored by tissue"); plt.xlabel("PC1"); plt.ylabel("PC2"); plt.tight_layout(); plt.show()

# t-SNE colored by tissue + centroids
plt.figure(figsize=(7.2, 6))
sc = plt.scatter(tsne_df["tSNE_1"], tsne_df["tSNE_2"], c=codes_tissue, s=10, alpha=0.8)
cent = tsne_df.groupby("tissue", as_index=False)[["tSNE_1","tSNE_2"]].mean()
plt.scatter(cent["tSNE_1"], cent["tSNE_2"], s=60, facecolors='white', edgecolors='black', linewidths=0.8, zorder=3)
for _, r in cent.iterrows():
    plt.text(r["tSNE_1"], r["tSNE_2"], str(r["tissue"]), fontsize=7, ha='center', va='center',
             bbox=dict(facecolor='white', alpha=0.6, edgecolor='none', pad=1.5))
proxies = [mlines.Line2D([], [], marker='o', linestyle='None') for _ in range(len(tissues.categories))]
for i, p in enumerate(proxies):
    p.set_markerfacecolor(sc.cmap(sc.norm(i))); p.set_markeredgecolor('black'); p.set_markersize(6)
plt.legend(proxies, list(tissues.categories), title="Tissue", loc="best", fontsize=8, frameon=True)
plt.title("t-SNE (2D) colored by tissue"); plt.xlabel("tSNE 1"); plt.ylabel("tSNE 2"); plt.tight_layout(); plt.show()

# PCA colored by k-means clusters
plt.figure(figsize=(7.2, 6))
plt.scatter(pca_df["PC1"], pca_df["PC2"], c=codes_km, s=10, alpha=0.8)
plt.title("PCA (2D) colored by k-means clusters"); plt.xlabel("PC1"); plt.ylabel("PC2"); plt.tight_layout(); plt.show()

# t-SNE colored by agglomerative clusters
plt.figure(figsize=(7.2, 6))
plt.scatter(tsne_df["tSNE_1"], tsne_df["tSNE_2"], c=codes_ag, s=10, alpha=0.8)
plt.title("t-SNE (2D) colored by agglomerative clusters"); plt.xlabel("tSNE 1"); plt.ylabel("tSNE 2"); plt.tight_layout(); plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.lines as mlines

from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering, Birch
from sklearn.mixture import GaussianMixture
from sklearn.metrics import adjusted_rand_score, confusion_matrix

# ----- align inputs -----
X = gtex_archs4_projection_filtered.T.copy()  # samples x LVs
X.index.name = "sample_id"

meta = gtex_meta.loc[:, ["SAMPID", "SMTS"]].copy()
meta["SAMPID"] = meta["SAMPID"].astype(str).str.strip()
common = X.index.intersection(meta["SAMPID"])
X = X.loc[common]
meta = meta.set_index("SAMPID").loc[common].reset_index().rename(columns={"SAMPID":"sample_id", "SMTS":"tissue"})
tissues = pd.Categorical(meta["tissue"])
y_true = tissues.codes
K = len(tissues.categories)

# ----- scale + PCA (for clustering) -----
scaler = StandardScaler()
Xs = scaler.fit_transform(X.values)
np.random.seed(123)
pca_full = PCA(n_components=min(50, Xs.shape[1], Xs.shape[0]-1), random_state=123)
Z = pca_full.fit_transform(Xs)
var80_n = max(2, (np.cumsum(pca_full.explained_variance_ratio_) >= 0.80).argmax() + 1)
Zk = Z[:, :var80_n]

# ----- t-SNE once (shared across plots) -----
from inspect import signature
tsne_kwargs = dict(n_components=2, perplexity=30, init="pca", random_state=123, verbose=1)
if "n_iter" in signature(TSNE).parameters: tsne_kwargs["n_iter"] = 1000
elif "max_iter" in signature(TSNE).parameters: tsne_kwargs["max_iter"] = 1000
if "learning_rate" in signature(TSNE).parameters: tsne_kwargs["learning_rate"] = 200
Y = TSNE(**tsne_kwargs).fit_transform(Xs)

# ----- global tissue palette (consistent across methods) -----
unique_tissues = sorted(tissues.categories.tolist())
palette = list(mpl.cm.tab20.colors) + list(mpl.cm.tab20b.colors) + list(mpl.cm.tab20c.colors)
if len(unique_tissues) > len(palette):
    extra = mpl.cm.hsv(np.linspace(0, 1, len(unique_tissues) - len(palette), endpoint=False))
    palette = palette + [tuple(c[:3]) for c in extra]
tissue_color = {t: palette[i] for i, t in enumerate(unique_tissues)}

# helper: map cluster -> majority tissue name (via confusion matrix)
def _map_clusters_to_tissue_names(y_true_codes, y_pred_clusters):
    cm = confusion_matrix(y_true_codes, y_pred_clusters)
    # rows: true tissues (code order matches tissues.categories), cols: clusters
    mapping = {}
    for j in range(cm.shape[1]):
        mapping[j] = tissues.categories[np.argmax(cm[:, j])]
    return mapping

# legend layout
legend_ncol = min(10, max(3, int(np.ceil(len(unique_tissues) / 3))))

# ======================= K-MEANS =======================
km = KMeans(n_clusters=K, n_init=20, random_state=123)
lab_km = km.fit_predict(Zk)
ari_km = adjusted_rand_score(y_true, lab_km)
map_km = _map_clusters_to_tissue_names(y_true, lab_km)
pred_tissue_km = np.array([map_km[c] for c in lab_km])
colors_km = [tissue_color[t] for t in pred_tissue_km]

fig, ax = plt.subplots(figsize=(9.2, 7.2))
ax.scatter(Y[:,0], Y[:,1], c=colors_km, s=12, alpha=0.9, edgecolors="white", linewidths=0.2)
cent_km = pd.DataFrame({"x":Y[:,0], "y":Y[:,1], "tissue":pred_tissue_km}).groupby("tissue")[["x","y"]].mean().reset_index()
ax.scatter(cent_km["x"], cent_km["y"], s=80, facecolors="white", edgecolors="black", linewidths=0.9, zorder=3)
for _, r in cent_km.iterrows():
    ax.text(r["x"], r["y"], str(r["tissue"]), fontsize=8, ha="center", va="center",
            bbox=dict(facecolor="white", alpha=0.65, edgecolor="none", pad=1.5))
proxies = [mlines.Line2D([], [], marker="o", linestyle="None", markersize=6,
                         markerfacecolor=tissue_color[t], markeredgecolor="black")
           for t in unique_tissues]
ax.legend(proxies, unique_tissues, title="Tissue",
          loc="upper center", bbox_to_anchor=(0.5, -0.16),
          ncol=legend_ncol, fontsize=8, frameon=True)
ax.set_title(f"k-means (ARI={ari_km:.3f})"); ax.set_xlabel("tSNE 1"); ax.set_ylabel("tSNE 2")
plt.subplots_adjust(bottom=0.28); plt.show()
print("k-means ARI:", round(ari_km, 3))

# =================== AGGLOMERATIVE (Ward) ===================
ag = AgglomerativeClustering(n_clusters=K, linkage="ward")
lab_ag = ag.fit_predict(Zk)
ari_ag = adjusted_rand_score(y_true, lab_ag)
map_ag = _map_clusters_to_tissue_names(y_true, lab_ag)
pred_tissue_ag = np.array([map_ag[c] for c in lab_ag])
colors_ag = [tissue_color[t] for t in pred_tissue_ag]

fig, ax = plt.subplots(figsize=(9.2, 7.2))
ax.scatter(Y[:,0], Y[:,1], c=colors_ag, s=12, alpha=0.9, edgecolors="white", linewidths=0.2)
cent_ag = pd.DataFrame({"x":Y[:,0], "y":Y[:,1], "tissue":pred_tissue_ag}).groupby("tissue")[["x","y"]].mean().reset_index()
ax.scatter(cent_ag["x"], cent_ag["y"], s=80, facecolors="white", edgecolors="black", linewidths=0.9, zorder=3)
for _, r in cent_ag.iterrows():
    ax.text(r["x"], r["y"], str(r["tissue"]), fontsize=8, ha="center", va="center",
            bbox=dict(facecolor="white", alpha=0.65, edgecolor="none", pad=1.5))
proxies = [mlines.Line2D([], [], marker="o", linestyle="None", markersize=6,
                         markerfacecolor=tissue_color[t], markeredgecolor="black")
           for t in unique_tissues]
ax.legend(proxies, unique_tissues, title="Tissue",
          loc="upper center", bbox_to_anchor=(0.5, -0.16),
          ncol=legend_ncol, fontsize=8, frameon=True)
ax.set_title(f"Agglomerative (ARI={ari_ag:.3f})"); ax.set_xlabel("tSNE 1"); ax.set_ylabel("tSNE 2")
plt.subplots_adjust(bottom=0.28); plt.show()
print("Agglomerative ARI:", round(ari_ag, 3))

# =================== SPECTRAL ===================
sc = SpectralClustering(n_clusters=K, affinity="nearest_neighbors", n_neighbors=15, assign_labels="kmeans", random_state=123)
lab_sc = sc.fit_predict(Zk)
ari_sc = adjusted_rand_score(y_true, lab_sc)
map_sc = _map_clusters_to_tissue_names(y_true, lab_sc)
pred_tissue_sc = np.array([map_sc[c] for c in lab_sc])
colors_sc = [tissue_color[t] for t in pred_tissue_sc]

fig, ax = plt.subplots(figsize=(9.2, 7.2))
ax.scatter(Y[:,0], Y[:,1], c=colors_sc, s=12, alpha=0.9, edgecolors="white", linewidths=0.2)
cent_sc = pd.DataFrame({"x":Y[:,0], "y":Y[:,1], "tissue":pred_tissue_sc}).groupby("tissue")[["x","y"]].mean().reset_index()
ax.scatter(cent_sc["x"], cent_sc["y"], s=80, facecolors="white", edgecolors="black", linewidths=0.9, zorder=3)
for _, r in cent_sc.iterrows():
    ax.text(r["x"], r["y"], str(r["tissue"]), fontsize=8, ha="center", va="center",
            bbox=dict(facecolor="white", alpha=0.65, edgecolor="none", pad=1.5))
proxies = [mlines.Line2D([], [], marker="o", linestyle="None", markersize=6,
                         markerfacecolor=tissue_color[t], markeredgecolor="black")
           for t in unique_tissues]
ax.legend(proxies, unique_tissues, title="Tissue",
          loc="upper center", bbox_to_anchor=(0.5, -0.16),
          ncol=legend_ncol, fontsize=8, frameon=True)
ax.set_title(f"Spectral (ARI={ari_sc:.3f})"); ax.set_xlabel("tSNE 1"); ax.set_ylabel("tSNE 2")
plt.subplots_adjust(bottom=0.28); plt.show()
print("Spectral ARI:", round(ari_sc, 3))

# =================== GAUSSIAN MIXTURE ===================
gm = GaussianMixture(n_components=K, covariance_type="full", random_state=123, n_init=5)
lab_gm = gm.fit_predict(Zk)
ari_gm = adjusted_rand_score(y_true, lab_gm)
map_gm = _map_clusters_to_tissue_names(y_true, lab_gm)
pred_tissue_gm = np.array([map_gm[c] for c in lab_gm])
colors_gm = [tissue_color[t] for t in pred_tissue_gm]

fig, ax = plt.subplots(figsize=(9.2, 7.2))
ax.scatter(Y[:,0], Y[:,1], c=colors_gm, s=12, alpha=0.9, edgecolors="white", linewidths=0.2)
cent_gm = pd.DataFrame({"x":Y[:,0], "y":Y[:,1], "tissue":pred_tissue_gm}).groupby("tissue")[["x","y"]].mean().reset_index()
ax.scatter(cent_gm["x"], cent_gm["y"], s=80, facecolors="white", edgecolors="black", linewidths=0.9, zorder=3)
for _, r in cent_gm.iterrows():
    ax.text(r["x"], r["y"], str(r["tissue"]), fontsize=8, ha="center", va="center",
            bbox=dict(facecolor="white", alpha=0.65, edgecolor="none", pad=1.5))
proxies = [mlines.Line2D([], [], marker="o", linestyle="None", markersize=6,
                         markerfacecolor=tissue_color[t], markeredgecolor="black")
           for t in unique_tissues]
ax.legend(proxies, unique_tissues, title="Tissue",
          loc="upper center", bbox_to_anchor=(0.5, -0.16),
          ncol=legend_ncol, fontsize=8, frameon=True)
ax.set_title(f"Gaussian Mixture (ARI={ari_gm:.3f})"); ax.set_xlabel("tSNE 1"); ax.set_ylabel("tSNE 2")
plt.subplots_adjust(bottom=0.28); plt.show()
print("Gaussian Mixture ARI:", round(ari_gm, 3))

# =================== BIRCH ===================
bh = Birch(n_clusters=K)
lab_bh = bh.fit_predict(Zk)
ari_bh = adjusted_rand_score(y_true, lab_bh)
map_bh = _map_clusters_to_tissue_names(y_true, lab_bh)
pred_tissue_bh = np.array([map_bh[c] for c in lab_bh])
colors_bh = [tissue_color[t] for t in pred_tissue_bh]

fig, ax = plt.subplots(figsize=(9.2, 7.2))
ax.scatter(Y[:,0], Y[:,1], c=colors_bh, s=12, alpha=0.9, edgecolors="white", linewidths=0.2)
cent_bh = pd.DataFrame({"x":Y[:,0], "y":Y[:,1], "tissue":pred_tissue_bh}).groupby("tissue")[["x","y"]].mean().reset_index()
ax.scatter(cent_bh["x"], cent_bh["y"], s=80, facecolors="white", edgecolors="black", linewidths=0.9, zorder=3)
for _, r in cent_bh.iterrows():
    ax.text(r["x"], r["y"], str(r["tissue"]), fontsize=8, ha="center", va="center",
            bbox=dict(facecolor="white", alpha=0.65, edgecolor="none", pad=1.5))
proxies = [mlines.Line2D([], [], marker="o", linestyle="None", markersize=6,
                         markerfacecolor=tissue_color[t], markeredgecolor="black")
           for t in unique_tissues]
ax.legend(proxies, unique_tissues, title="Tissue",
          loc="upper center", bbox_to_anchor=(0.5, -0.16),
          ncol=legend_ncol, fontsize=8, frameon=True)
ax.set_title(f"Birch (ARI={ari_bh:.3f})"); ax.set_xlabel("tSNE 1"); ax.set_ylabel("tSNE 2")
plt.subplots_adjust(bottom=0.28); plt.show()
print("Birch ARI:", round(ari_bh, 3))
```
